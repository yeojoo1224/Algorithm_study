```python
#거스름돈 문제

n= int(input()) # 거슬러주어야 할 돈
a = 0
b = 0
c = 0
d= 0

a= n//500
b = (n - 500*a) //100
c = (n-500*a-100*b)//50
d = (n-500*a-100*b-50*c)//10

print(a+b+c+d, "개")
```

```python
#거스름돈 문제 모범답안
n = int(input()) #1260
cnt = 0

coin_types = [500,100,50,10]

for i in coin_types:
  cnt += n//i #1260//500, 1260//100 #cnt = 해당 동전 '몇 개'인지 알기
  n = n%i #n = 처음 나눈수의 나머지로 변환

print(cnt)
```
