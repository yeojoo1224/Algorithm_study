```python
#숫자 카드 게임
n, m = map(int, input().split())
min = 0

for i in range(m):
  num = list(map(int,input().split()))
  num.sort() #오름차순 정렬 
  if(num[0]>min): #최소보다 크면
    min = num[0]  #최소값을 min에 집어넣음

print(min)
```
